```python
%matplotlib widget
import numpy as np
import obspy
import matplotlib.pyplot as plt
from matplotlib.widgets import Button
import matplotlib.patches as mpatches
import matplotlib.dates as mdates
```

```python
# set path and data parameters
path = "/media/Data/Data/PIG/"
stat = "PIG*"
chans = "HH*"

# set day of interest
year = "2012"
month = "05"
day = "09"

# read in traces without instrument response removed
files = path + "MSEED/raw/" + stat + "/" + chans + "/*" + chans + "__" + year + month + day + "T000000Z__" + "*"
```

```python
# read in raw data
st_raw = obspy.read(files)

# basic preprocessing
st_raw.detrend("demean")
st_raw.detrend("linear")
st_raw.taper(max_percentage=0.00025, max_length=20.)

# set path to XML response file
pathXML = path + "XML/" + chans + "/*" + stat + ".xml"
inv = obspy.read_inventory(pathXML)

# remove response
st_100s = st_raw.copy()
st_1000s = st_raw.copy()
st_100s.remove_response(inventory=inv,pre_filt=[0.005,0.01,45,50],output="VEL")
st_1000s.remove_response(inventory=inv,pre_filt=[0.0005,0.001,45,50],output="VEL")

# save outputs
st_raw.write("may_9_velocity_raw.MSEED")
st_100s.write("may_9_velocity_100s.MSEED")
st_1000s.write("may_9_velocity_1000s.MSEED")
```

```python
# get data
st_raw = obspy.read("may_9_velocity_raw.MSEED")
st_100s = obspy.read("may_9_velocity_100s.MSEED")
st_1000s = obspy.read("may_9_velocity_1000s.MSEED")

# downsample so plotting isn't horribly slow
st_raw.resample(1)
st_100s.resample(1)
st_1000s.resample(1)

# get response
pathXML = path + "XML/" + chans + "/*" + stat + ".xml"
inv = obspy.read_inventory(pathXML)
```

```python
class metadata:
    station = "PIG2"
    channel = "HHZ"
    
    def select_channel(self,channel):
        def clicked(event):
            self.channel = channel
            
            # plot each trace
            plot_raw.set_ydata(st_raw.select(station=self.station,channel=self.channel)[0].data[:trace_length]*1000)
            plot_100s.set_ydata(st_100s.select(station=self.station,channel=self.channel)[0].data[:trace_length]*1000)
            plot_1000s.set_ydata(st_1000s.select(station=self.station,channel=self.channel)[0].data[:trace_length]*1000)

            ax[0][0].set(title=self.station+" "+self.channel+"\nInstrument Response")
            for axis in ax[1:]:
                axis.relim()
                axis.autoscale()
                axis.set_xlim([t[0],t[-1]])
            plt.draw()
        return clicked
    
    def select_station(self,station):
        def clicked(event):
            self.station = station
            plot_raw.set_ydata(st_raw.select(station=self.station,channel=self.channel)[0].data[:trace_length]*1000)
            plot_100s.set_ydata(st_100s.select(station=self.station,channel=self.channel)[0].data[:trace_length]*1000)
            plot_1000s.set_ydata(st_1000s.select(station=self.station,channel=self.channel)[0].data[:trace_length]*1000)
            
            ax[0][0].set(title=self.station+" "+self.channel+"\nInstrument Response")
            for axis in ax[1:]:
                axis.relim()
                axis.autoscale()
                axis.set_xlim([t[0],t[-1]])
            plt.draw()
        return clicked

# make plot
fig,ax = plt.subplots(4,3,figsize=[20,10])
plt.subplots_adjust(bottom=0.2)
colors = plt.rcParams['axes.prop_cycle'].by_key()['color']

# get data and plot
trace_length = 86400
t = st_raw[1].times(type="matplotlib")[:trace_length]
data_raw = st_raw.select(station="PIG2",channel="HHZ")[0].data[:trace_length]*1000
data_100s = st_100s.select(station="PIG2",channel="HHZ")[0].data[:trace_length]*1000
data_1000s = st_1000s.select(station="PIG2",channel="HHZ")[0].data[:trace_length]*1000
plot_raw, = ax[1][0].plot(t,data_raw, lw=2,color=colors[0])
plot_100s, = ax[2][0].plot(t,data_100s, lw=2,color=colors[1])
plot_1000s, = ax[3][0].plot(t,data_1000s, lw=2,color=colors[2])

# plot instrumental response
sampling_rate = 100
min_freq = 0.0001
t_samp = 1.0 / sampling_rate
nyquist = sampling_rate / 2.0
nfft = int(sampling_rate / min_freq)
response = inv.select(channel="HHZ",station="PIG2")[0][0][0].response
cpx_response, freq = response.get_evalresp_response(t_samp=t_samp, nfft=nfft)
plot_resp, = ax[0][0].loglog(1/freq[freq<nyquist],abs(cpx_response[freq<nyquist]),'k')
ax[0][0].invert_xaxis()
callback = metadata()

# configure labels
ax[0][0].set_ylabel("Amplitiude")
ax[0][0].set_xlabel("Period (s)")
ax[1][0].set_ylabel("Counts")
ax[1][0].set_title("Raw",y=0.8)#,x=0.95)
ax[2][0].set_ylabel("Velocity (mm/s)")
ax[2][0].set_title(">100s",y=0.8)#,x=0.95)
ax[3][0].set_ylabel("Veloctiy (mm/s)")
ax[3][0].set_title(">1000s",y=0.8)#,x=0.95)
ax[3][0].set_xlabel("Time")
ax[3][0].xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
for axis in ax[1:][0]:
    axis.set_xticks([t[0],t[len(t)//4],t[len(t)//2],t[len(t)*3//4],t[-1]])
    axis.set_xlim([t[0],t[-1]])
ax[1][0].get_xaxis().set_ticklabels([])
ax[2][0].get_xaxis().set_ticklabels([])
ax[3][0].get_shared_x_axes().join(ax[3][0],ax[1][0])
#ax[3][0].get_shared_x_axes().join(ax[3][0],ax[2][0])
box = ax[0][0].get_position()
box.y0 = box.y0 + 0.05
box.y1 = box.y1 + 0.05
ax[0][0].set_position(box)

axbutton = plt.axes([0.125, 0.1, 0.05, 0.05])
bpig2 = Button(axbutton, 'PIG2')
bpig2.on_clicked(callback.select_station("PIG2"))

axbutton = plt.axes([0.2, 0.1, 0.05, 0.05])
bpig4 = Button(axbutton, 'PIG4')
bpig4.on_clicked(callback.select_station("PIG4"))

axbutton = plt.axes([0.275, 0.1, 0.05, 0.05])
bpig5 = Button(axbutton, 'PIG5')
bpig5.on_clicked(callback.select_station("PIG5"))
plt.text(-1.6,1.125,"Stations")

# station selection buttons
axbutton = plt.axes([0.125, 0.01, 0.05, 0.05])
bhhz = Button(axbutton, 'HHZ')
bhhz.on_clicked(callback.select_channel("HHZ"))

axbutton = plt.axes([0.2, 0.01, 0.05, 0.05])
bhhn = Button(axbutton, 'HHN')
bhhn.on_clicked(callback.select_channel("HHN"))

axbutton = plt.axes([0.275, 0.01, 0.05, 0.05])
bhhe = Button(axbutton, 'HHE')
bhhe.on_clicked(callback.select_channel("HHE"))
plt.text(-1.6,1.125,"Channels")

plt.show()
```
